```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load with correct separator
df = pd.read_csv('student-por.csv', sep=',')

# First look at the data
print("Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
# Step 2 - Data Cleaning
print("Missing Values in each column:")
print(df.isnull().sum())

print("\nData Types:")
print(df.dtypes)

print("\nBasic Statistics:")
df.describe()
```

```python
# Step 3 - EDA: Grade Distribution
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.hist(df['G1'], bins=15, color='steelblue', edgecolor='black')
plt.title('G1 - First Period Grades')
plt.xlabel('Grade')

plt.subplot(1, 3, 2)
plt.hist(df['G2'], bins=15, color='orange', edgecolor='black')
plt.title('G2 - Second Period Grades')
plt.xlabel('Grade')

plt.subplot(1, 3, 3)
plt.hist(df['G3'], bins=15, color='green', edgecolor='black')
plt.title('G3 - Final Grades')
plt.xlabel('Grade')

plt.suptitle('Distribution of Student Grades', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Step 4 - Correlation Heatmap
plt.figure(figsize=(12, 8))

# Select only numeric columns
numeric_df = df[['age', 'studytime', 'failures', 'famrel',
                  'freetime', 'goout', 'absences', 'G1', 'G2', 'G3']]

sns.heatmap(numeric_df.corr(),
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            square=True)

plt.title('Correlation Between Student Performance Factors', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Step 5 - Scatter Plot: Study Time vs Final Grade
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='studytime', y='G3', alpha=0.6, color='steelblue')
sns.regplot(data=df, x='studytime', y='G3', scatter=False, color='red')
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time (1=low, 4=high)')
plt.ylabel('Final Grade (G3)')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='absences', y='G3', alpha=0.6, color='orange')
sns.regplot(data=df, x='absences', y='G3', scatter=False, color='red')
plt.title('Absences vs Final Grade')
plt.xlabel('Number of Absences')
plt.ylabel('Final Grade (G3)')

plt.tight_layout()
plt.show()
```

```python
# Step 6 - Final Grade by Number of Failures + Internet Access
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Average grade by failures
failure_grades = df.groupby('failures')['G3'].mean().reset_index()
axes[0].bar(failure_grades['failures'], failure_grades['G3'],
            color=['green','orange','red','darkred'])
axes[0].set_title('Average Final Grade by Past Failures')
axes[0].set_xlabel('Number of Past Failures')
axes[0].set_ylabel('Average Final Grade')
for i, v in enumerate(failure_grades['G3']):
    axes[0].text(i, v + 0.2, f'{v:.1f}', ha='center', fontweight='bold')

# Average grade by internet access
internet_grades = df.groupby('internet')['G3'].mean().reset_index()
axes[1].bar(internet_grades['internet'], internet_grades['G3'],
            color=['steelblue', 'coral'], width=0.4)
axes[1].set_title('Average Final Grade by Internet Access')
axes[1].set_xlabel('Has Internet Access (no/yes)')
axes[1].set_ylabel('Average Final Grade')
for i, v in enumerate(internet_grades['G3']):
    axes[1].text(i, v + 0.2, f'{v:.1f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Step 7 - Final Insights Summary
print("=" * 55)
print("   STUDENT PERFORMANCE ANALYSIS - KEY INSIGHTS")
print("=" * 55)

avg_g3 = df['G3'].mean()
print(f"\n📊 Dataset: {df.shape[0]} students, {df.shape[1]} features")
print(f"📈 Average Final Grade (G3): {avg_g3:.2f} / 20")

# Insight 1: Study time impact
low_study = df[df['studytime'] <= 2]['G3'].mean()
high_study = df[df['studytime'] >= 3]['G3'].mean()
diff1 = ((high_study - low_study) / low_study) * 100
print(f"\n✅ Insight 1 - Study Time Impact:")
print(f"   Low study time avg grade  : {low_study:.2f}")
print(f"   High study time avg grade : {high_study:.2f}")
print(f"   Higher study = {diff1:.1f}% better grades")

# Insight 2: Failures impact
no_fail = df[df['failures'] == 0]['G3'].mean()
has_fail = df[df['failures'] > 0]['G3'].mean()
diff2 = ((no_fail - has_fail) / has_fail) * 100
print(f"\n✅ Insight 2 - Past Failures Impact:")
print(f"   Students with 0 failures  : {no_fail:.2f} avg")
print(f"   Students with failures    : {has_fail:.2f} avg")
print(f"   No failures = {diff2:.1f}% higher grades")

# Insight 3: Internet access
no_net = df[df['internet'] == 'no']['G3'].mean()
yes_net = df[df['internet'] == 'yes']['G3'].mean()
diff3 = ((yes_net - no_net) / no_net) * 100
print(f"\n✅ Insight 3 - Internet Access Impact:")
print(f"   Without internet avg grade: {no_net:.2f}")
print(f"   With internet avg grade   : {yes_net:.2f}")
print(f"   Internet access = {diff3:.1f}% better grades")

print("\n" + "=" * 55)
print("   Analysis Complete!")
print("=" * 55)
```
